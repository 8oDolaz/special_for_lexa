```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharex=True, sharey=True)
x = np.arange(-1/4,1/4*np.pi,0.1)
ax1.plot(x, np.abs(np.cos(x*x-3)))
ax1.plot(x, (np.sin(np.sqrt(x))))
ax2.plot(x, (np.power(x, 4)+7*np.abs(np.power(x, 3))-2))

plt.show()
```
